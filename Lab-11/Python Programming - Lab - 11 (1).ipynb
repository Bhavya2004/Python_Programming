### 01) WAP to demonstrate the use of Pie chart.

```python
fruits = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes = [30, 25, 20, 25]
ex = [0, 0.1, 0, 0]  # 'explode' the second slice (i.e. 'Bananas')
color=['r','g','y','m'] #change the color

plt.pie(sizes, explode=ex, labels=fruits, autopct='%1.1f%%',colors=color,shadow=True,radius=1.2,labeldistance=1.2,textprops={"fontsize":10,"color":'b'})
# plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Fruit Consumption')

plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
values = [30, 25, 20, 15]
color=['r','g','y','m']

# Create a bar chart
plt.bar(labels, values,color=color,align='edge',edgecolor='b',alpha=0.5)

# Add labels to the X and Y axis and a title to the chart
plt.xlabel('Fruits')
plt.ylabel('Quantity')
plt.title('Fruit Inventory')

# Display the chart
plt.show()
```

```python
#horizontal 

labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
values = [30, 25, 20, 15]
color=['r','g','y','m']

# Create a bar chart
plt.barh(labels, values,color=color,align='edge',edgecolor='b',alpha=0.5)

# Add labels to the X and Y axis and a title to the chart
plt.xlabel('Fruits')
plt.ylabel('Quantity')
plt.title('Fruit Inventory')

# Display the chart
plt.show()
```

### 04) WAP to demonstrate the use of Histogram.

```python
# Generate a list of 100 random values between 0 and 50
data = [random.randint(0, 50) for i in range(100)]

# Create a histogram
plt.hist(data)

# Add labels to the X and Y axis and a title to the chart
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Random Data Histogram')

# Display the chart
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x = ['Apples', 'Bananas', 'Oranges', 'Grapes']
y = [30, 25, 20, 15]
color=['r','g','y','m']

# Create a bar chart
plt.bar(labels, values,color=color,align='edge',edgecolor='b',alpha=0.5)

# Add labels to the X and Y axis and a title to the chart
plt.xlabel('Fruits')
plt.ylabel('Quantity')
plt.title('Fruit Inventory')
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
addlabels(x,y)
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
c=['r','b','g','y','c']
size=[50,75,40,25,80]
plt.scatter(x,y,color=c,marker='*',s=size,edgecolor='b')
plt.xlabel("Age")
plt.ylabel("Height")
plt.title("Scatter Plot")
plt.show()
```
